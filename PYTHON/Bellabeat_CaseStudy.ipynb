### 3.1 Dataset Used:<a class="anchor" id="dataset_used_3_1"></a>  
* The data source used for our case study is FitBit Fitness Tracker Data.
* The data is publicly available on [Kaggle: FitBit Fitness Tracker Data](https://www.kaggle.com/datasets/nurudeenabdulsalaam/fitbit-fitness-tracker-data) and stored in 18 csv files. 
* These datasets were generated by respondents from a distributed survey via Amazon Mechanical Turk between 12 March 2016 to 12 May 2016.
* 30 FitBit users consented to the submission of personal tracker data.
* Data collected includes (1) physical activity recorded in minutes, (2) heart rate, (3) sleep monitoring, (4) daily activity and (5) steps.
* Variation between output represents use of different types of Fitbit trackers and individual tracking behaviors / preferences. 

### 3.2 Dataset Limitations:<a class="anchor" id="dataset_limitations_3_2"></a> 

* Data collected from year 2016 which is more than the standard time lapse. Users' daily activity, fitness and sleeping habits, diet and food consumption may have changed since then, hence data may not be relevant.
* Sample size of 30 female FitBit users is not representative of the entire female population so we could encounter a sampling bias. 
* As data is collected in a survey, hence unable to ascertain the integrity or accuracy of data.

In summary, the dataset is is of low quality data and it is not advisable to utilize it in generating business recommendations.

### 5.2.1 Distribution of Users by Activity Levels<a class="anchor" id="distribution_of_users_by_activity_levels_5_2_1"></a>   

According to a study done by the [BioMedical Central (BMC)](https://ijbnpa.biomedcentral.com/articles/10.1186/1479-5868-8-79), healthy adults typically take between 4,000 and 18,000 steps/day, and that 10,000 steps/day is reasonable for a healthy population. We will therefore be categorizing users by considering their daily amount of steps. Categorization will be done based on the [Pedometer Indices created by The 10,000 Steps Organization](https://www.10000steps.org.au/articles/healthy-lifestyles/counting-steps/).

* Sedentary = less than 5,000 steps per day 
* Low active = 5,000 to 7,499 steps per day
* Fairly active = 7,500 to 9,999 steps per day
* Very Active = more than 10,000 steps per day
* Highly active = more than 12,500 

```python
# Calculating the average daily steps per user

user_da = activitysleep_da.groupby(['Id']).mean()
user_da.head()
```

```python
# Creating new column UserCategory based on TotalSteps
def conditions(x):
    if   x < 5000:   return "Sedentary"
    elif x <= 7499: return "Low Active"
    elif x >= 12500: return "Highly Active"
    elif x > 10000: return "Very Active"
    else:           return "Fairly Active"

func         = np.vectorize(conditions)
UserCategory = func(user_da["TotalSteps"])

#Adding column in dataframe
user_da["UserCategory"] = UserCategory

user_da[["TotalSteps","UserCategory"]].tail()
```

```python
# Creating UserCategory dataframe
UserCategory_da = user_da.groupby(['UserCategory']).size()

UserCategory_da.nunique
```

```python
# Creating dataset
UserCategory = ['Fairly Active', 'Highly Active', 'Very Active',
        'Sedentary', 'Low Active']
 
data = [10, 3, 9, 7, 4]

#Creating explode data
explode = (0.1, 0.2, 0.0, 0.0, 0.1)
 
# Creating color parameters
colors = ( 'orange', 'cyan', 'brown',
          'grey', 'indigo')
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : 'green' }
 
# Creating autopct arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return '{:.1f}%'.format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(8, 5))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = UserCategory,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="black"))
 
# Adding legend
ax.legend(wedges, UserCategory,
          title ='User Activity Levels',
          loc ='center right',
          bbox_to_anchor =(1.1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 10, weight ='bold')
ax.set_title('Distribution of Users by Activity Levels')
 
# show plot
plt.show()
```

#### *Distribution of Users by Activity Levels*

According to research we need users to fall on a minimum into the **"Fairly Active"** category as this is the recommended daily step target for healthy adults to experience maximum health benefits. Fortunately, we have the greatest spread of users in the Fairly Active category, however, the Highly Active category records the lowest spread of users. 


### 5.2.2 Distribution of Users by Usage Frequency<a class="anchor" id="distribution_of_users_by_usage_frequency_5_2_2"></a>

We will be categorizing users based on how often they use their FitBit App. We will be assuming that days with less than 200 TotalSteps taken, are days where users have not used their App. We will filter out these inactive days and assign the following designations:

* Low Use - 1 to 14 days
* Moderate Use - 15 to 21 days
* High Use - 22 to 31 days



```python
# Counting the number of days that the App is in use per user

user_df = activitysleep_da.groupby(['Id']).count()
user_df.head()
```

```python
# Renaming the column "ActivityDate" to "ActivityDays"
user_df.rename(columns = {'ActivityDate':'ActivityDays'}, inplace = True)

user_df.head(3)
```

```python
# Creating new column UserFreq based on ActivityDate
def conditions(z):
    if   z < 14: return "Low Use"
    elif z > 22: return "High Use"
    else:        return "Moderate Use"

func         = np.vectorize(conditions)
UserFreq = func(user_df["ActivityDays"])

#Adding column in dataframe
user_df["UserFreq"] = UserFreq

user_df[["ActivityDays","UserFreq"]].head()
```

```python
# Creating UserFreq dataframe
UserFreq_da = user_df.groupby(['UserFreq']).size()

UserFreq_da.nunique
```

```python
# Creating dataset
UserFreq = ['High Use', 'Low Use', 'Moderate Use']
 
data = [21, 1, 8]

#Creating explode data
explode = (0.1, 0.0, 0.0)
 
# Creating color parameters
colors = ( 'orange', 'brown', 'indigo')
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : 'green' }
 
# Creating autopct arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return '{:.1f}%'.format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(6, 3))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = UserFreq,
                                  shadow = False,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="black"))
 
# Adding legend
ax.legend(wedges, UserFreq,
          title ='Usage Frequency',
          loc ='center right',
          bbox_to_anchor =(1.1, 0, 0.8, 1))
 
plt.setp(autotexts, size = 8, weight ='bold')
ax.set_title('Distribution of Users by Usage Frequency')
 
# show plot
plt.show()
```

#### *Distribution of Users by Usage Frequency*

Analyzing our results we can see that

* 70% of our sample use their device frequently i.e, for 22 days and more

### 6.1 Summary: Trends and Insights<a class="anchor" id="summary_trends_and_insights_6_1"></a> 

**Device Usage**

More users used the device to count their steps than users who also used it for measuring sleep patterns, weight etc. Of a total of thirty users, 33 unique Ids used daily and hourly step count function, compared to that of 24 unique IDs for sleep tracking, and only 8 of them used the device to manage their weight. This could be because these users have other devices they also use or maybe more than one unique ID.  However, go by the insight gained by the [CCS Insight](https://www.ccsinsight.com/press/company-news/1957-clear-marketing-needed-to-drive-sales-of-smartwatches-and-fitness-trackers/) into the use of smartwatches, stating that "49% of people do not understand the benefits of smartwatches and that some of the functions appeal to them", which may mean that some functions do not appeal to the users, the functions of sleep and weight tracking may fall into this non-appealing category.




**Usage Frequency**

The 'High Use' group consists of 24 users or 70% of the total sample size. They use the device very frequently, between 22-31 days. They are the most active group and so unsurprisingly, they burn the most calories (almost 3000 calories). They also get more sleep than the other two categories.
The 'Moderate Use' group however, consists of 8 users or 26.7% of the total sample size. They use the device between 15 - 21 days. Users in this group are less active, walking fewer steps compared to the High Use group, their calories burnt is way less than that of the 'High Use' group, with less than 2500 maximum calories burnt, they are lower than the median calories burnt for the 'High Use' group, however, their sleep time seem to be almost at par with that of the 'High Use' group
Finally, the 'Low Use' group consist of only 1 user or 3.3% of our total sample size. This sample size is totally inadequate to provide any meaningful insight or recognizable patterns. Knowing this, this group use their devices for less than 14 days and display the lowest activity levels and calories burnt of all 3 groups. 



**User's Daily Activity**

Based on our summary statistics, although the average user spends a little over 247 minutes in either of lightly active, fairly active or very active positions, 79.4% of their time is actually spent in a sedentary position.



**User's Statistics**

The average user takes 8,329 steps and burns 2,362 calories per day. These average steps taken are lower than the daily 10,000 steps recommended by CDC. However our users are getting the recommended amount of sleep according to the Sleep Foundation, with an average sleep time of 419 minutes (7 hours)



**User category**

The Fairly Active user category records the greatest spread of users, with Highly Active recording the lowest. However, the 'Highly Active' category burns the most calories followed by the fairly active with the "Low Active" and the "Very Active" categories been almost at par in the maximum calories they burn daily.



**Activity Levels**

Activity levels during the week drops on Wednesday and decreases consistently till the weekends, with Tuesdays been the most active day and Sunday the least.
Users display higher activity levels in the mornings from 6am till 7pm, the activity levels stays consistently above 500 steps per hour up until 19:00 hours when there is a sharp drop and consistent decline. The most active periods are shown to be between 5pm - 7pm.
There is a slightly weak positive correlation between Activity Levels and Usage Frequency.



**Calories Burnt**

Calories burnt during the weekdays drops from Thursday, and decreases consistently through the weekend till Tuesday in the following week where it picks up again, with Wednesdays recording the most calories burnt and Sundays the least.
The amount of calories burnt by a user steadily increases from the moment the user wakes up, peaking around at the same times that the users display the most activity during the day (5pm-7pm).
There is a positive relationship between the measures of calories burnt and users active days as well as between calories burnt and activity levels. Although relationship between calories burnt and users' active days is quite weak.



**Sleep Time**

Sundays are most likely rest days, seeing as Sunday records the lowest number of steps taken, the least calories burnt and the highest number of minutes asleep.
There is a strong, positive relationship between the total number of minutes asleep and the total time spent in bed. On the average users spend about 39 minutes extra in bed before going to sleep.
